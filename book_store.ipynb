```python
import chromedriver_autoinstaller
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from pymongo import MongoClient
from IPython.display import HTML
import pandas as pd
```

```python
def listAppend(img_src, title, author, publish, created, price, sales_num = 'x'):
    img_list.append(img_src)
    title_list.append(title)
    author_list.append(author)
    publish_list.append(publish)
    created_list.append(created)
    price_list.append(price)
    sales_num_list.append(sales_num)
    return True
```

## 예스24

```python
def yes24(keyword, categoryNum):
    driver = webdriver.Chrome()
    url = f'https://www.yes24.com/Product/Search?domain=ALL&query={keyword}'
    driver.get(url)

    if categoryNum != '1':
        category = driver.find_element("xpath", f"/html/body/div[1]/div[4]/div/div[2]/section[2]/div[3]/div/span[1]/a[{categoryNum}]").click()

    for i in range(20):
        driver.execute_script(f'window.scrollTo(0, document.body.scrollHeight / 20 * {i})')
        time.sleep(0.1)

    li_elements_xpath = '/html/body/div[1]/div[4]/div/div[2]/section[2]/div[3]/ul/li'
    li_elements = driver.find_elements(By.XPATH, li_elements_xpath)
    
    for li in li_elements:
        try:
            img = li.find_element(By.TAG_NAME, 'img')
            title = li.find_element(By.CLASS_NAME,'gd_name')
            try:
                author = li.find_element(By.CLASS_NAME, 'info_auth').text.split(' 저')[0]
            except:
                author = "정보 없음"
            try:
                publish = li.find_element(By.CLASS_NAME, 'info_pub').text
            except:
                publish = "정보 없음"
            created = li.find_element(By.CLASS_NAME, 'info_date').text
            price = li.find_element(By.CLASS_NAME, 'yes_b').text
            sales_num = li.find_elements(By.CLASS_NAME, 'saleNum')
            if sales_num:
                sales_num = sales_num[0].text.split(' ')[1]
            else:
                sales_num = '0'
            
            title = img.get_attribute('alt')
            img_src = img.get_attribute('src')
            listAppend(img_src, title, author, publish, created, price, sales_num)
            
        except:
            print('오류')
            continue  # 이미지가 없는 경우 건너뜀
```

## 인터파크

```python
def interpark(keyword, categoryNum):
    driver = webdriver.Chrome()
    url = 'https://book.interpark.com/bookPark/html/book.html'
    driver.get(url)

    search = driver.find_element("xpath", "/html/body/div[2]/div[3]/div[2]/div/div/div[1]/form/div[1]/input[9]")
    search.send_keys(keyword)
    button = driver.find_element("xpath", "/html/body/div[2]/div[3]/div[2]/div/div/div[1]/form/div[2]/a").click()

    if categoryNum != '1':
        categroy = driver.find_element("xpath", f"/html/body/div[3]/div[2]/span[1]/ul/li[{categoryNum}]/a").click()

    for i in range(20):
        driver.execute_script(f'window.scrollTo(0, document.body.scrollHeight / 20 * {i})')
        time.sleep(0.1)
    
    li_elements_xpath = '/html/body/div[3]/div[2]/span[1]/form'
    li_elements = driver.find_elements(By.CLASS_NAME, 'list_wrap')
        
    for li in li_elements:
        try:
            img = li.find_element(By.CLASS_NAME, 'bd')
            title = li.find_element(By.TAG_NAME,'b')
            w_list = li.find_element(By.CLASS_NAME, 'writer').text.split(' 저')
            elements = li.find_elements(By.CLASS_NAME, 'nowMoney')
            if elements:
                price = elements[0].text.split('원')[0]
            else:
                price = li.find_element(By.CLASS_NAME, 'defaultNum').text.split('원')[0]
            sales_num = li.find_elements(By.CLASS_NAME, 'saleIndex')
            if sales_num:
                sales_num = sales_num[0].text.split(' ')[1] # 판매지수
            else:
                sales_num = '0'
    
            img_src = img.get_attribute('src')
            title = title.text
            if len(w_list)!=1:
                author = w_list[0]
                publish = w_list[1].split('I')[-2]
                created = w_list[1].split('I')[-1]
            else:
                author = 'X'
                publish = w_list[0].split('I')[-2]
                created = w_list[0].split('I')[-1]
                
            listAppend(img_src, title, author, publish, created, price, sales_num)
            
        except:
            print('오류')
            continue  # 이미지가 없는 경우 건너뜀
```

```python
def aladin(keyword, categoryNum):
    driver = webdriver.Chrome()
    url = f'https://www.aladin.co.kr/m/msearch.aspx?SearchWord={keyword}&SearchTarget=All'   
    driver.get(url)
    if categoryNum != '1':
        category = driver.find_element("xpath", f"/html/body/div[8]/div[2]/div[1]/select/option[{categoryNum}]").click()
    
    for i in range(20):
        driver.execute_script(f'window.scrollTo(0, document.body.scrollHeight / 20 * {i})')
        time.sleep(0.1)
    
    li_elements_xpath = '/html/body/div[12]'
    li_elements = driver.find_elements(By.CLASS_NAME, 'browse_list_box')
        
    for li in li_elements:
        try:
            img = li.find_elements(By.CLASS_NAME, 'front_cover')
            if img:
                img_src = img[0].get_attribute('src')
            else:
                img_src = li.find_element(By.CLASS_NAME, 'i_cover').get_attribute('src')
                
            title = li.find_element(By.CLASS_NAME,'b_book_t').text.split('] ')[1]
            author = li.find_elements(By.TAG_NAME, 'li')[1].text.split(' (지은이)')[0]
            
            elements = li.find_elements(By.TAG_NAME, 'li')[2].text
            publish = elements.split(' | ')[0]
            created = elements.split(' | ')[1]
    
            price = li.find_element(By.CLASS_NAME, 'b_price').text
            sales_num = 'x'
                
            listAppend(img_src, title, author, publish, created, price, sales_num)
        except:
            print('오류')
            continue  # 이미지가 없는 경우 건너뜀
```

```python
def kyobo(keyword, categoryNum):
    driver = webdriver.Chrome()
    categoryNum = int(categoryNum)
    if (categoryNum == 1):
        url = f"https://search.kyobobook.co.kr/search?keyword={keyword}&gbCode=TOT&target=total"
    else:
        categoryName = ["qntt", "date", "name", "dc", "prasc", "prdesc", "kcont", "krvgr"]
        url = f"https://search.kyobobook.co.kr/search?keyword={keyword}&target=total&gbCode=TOT&ra={categoryName[categoryNum - 2]}"
    driver.get(url)
    time.sleep(2)
    # ▲ 검색
    
    for i in range(20):
        driver.execute_script(f"window.scrollTo(0, 1000 * {i})")
        time.sleep(0.1)
    # ▲ 스크롤 

    li_element_xpath = "/html/body/div[3]/main/section/div/div/div[4]/div[2]/div/div[2]/div[3]/ul"
    li_element_xpath_load = driver.find_element(By.XPATH, li_element_xpath)
    li_elements = li_element_xpath_load.find_elements(By.CLASS_NAME, "prod_item")


    for li in li_elements:
        try:
            img = li.find_element(By.CLASS_NAME, "prod_img_load")
            title = li.find_element(By.CLASS_NAME, "prod_info").text.split(']')[1]
            author = li.find_element(By.CLASS_NAME, "author.rep").text
            publish = li.find_element(By.CLASS_NAME, "prod_publish").text.split("\n")[0]
            created = li.find_element(By.CLASS_NAME, "prod_publish").text.split("\n")[1]
            try:
                price = li.find_element(By.CLASS_NAME, "price").text
            except:
                price = li.find_element(By.CLASS_NAME, "prod_purchase_state").text
            sales_num = 'x'

            img_src = img.get_attribute("src")

            listAppend(img_src, title, author, publish, created, price, sales_num)
        except:
            print("오류")
            continue
```

```python
def resetCollectionList():
    global collection, img_list, title_list, author_list, publish_list, created_list, price_list, sales_num_list
    collection = ''
    img_list = []
    title_list = []
    author_list = []
    publish_list = []
    created_list = []
    price_list = []
    sales_num_list = []
```

```python
def start():
    print('1.yes24 | 2.interpark | 3.aladin | 4.kyobo')
    num = input("사이트 선택: ")
    search(num)
    dataTot = []
    for i in range(len(img_list)):
        url = img_list[i]
        name = title_list[i]
        author = author_list[i]
        publisher = publish_list[i]
        publicationDate = created_list[i]
        price = price_list[i]
        sales_num = sales_num_list[i]
        dataTot.append([url, name, author, publisher, publicationDate, price, sales_num])

    col = ["이미지", "이름", "저자", "출판사", "출판일", "가격", "판매량"]
    df = pd.DataFrame(data=dataTot, columns=col)
    # ▲ 데이터프레임으로 저장

    def image_html(url):
        return '<img src="' + url + '" width="60">'
    # ▲ 이미지를 HTML태그로 변환하는 함
    
    df["이미지"] = df["이미지"].apply(image_html)
    # ▲ HTML 태그 적용
    
    show = display(HTML(df.to_html(escape=False)))
    # ▲ HTML로 랜더링 되도록 설정? < 이 부분을 잘 모르겠음.
    
    return show
```
